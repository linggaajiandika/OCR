### OCR

```
img_cv = cv2.imread(r'/content/img_20190511_084303.jpg')
d = cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB)
```

```
from matplotlib import pyplot as plt
plt.imshow(d, interpolation='nearest')
plt.show()
```

```
print(pytesseract.image_to_string(d, lang='eng', config='-psm 1'))
```
